```python
import requests

def download_pic(url, path):
    pic = requests.get(url)
    path += url[url.rfind("."):]
    f = open(path, "wb")
    f.write(pic.content)
    f.close()

url = "https://live.staticflickr.com/65535/48414373092_1402e4bdb9_b.jpg"
pic_path = "download"
download_pic(url, pic_path)
```

```python
import threading

def job(num):
    print("子執行緒", num)
    
threads = []
for i in range(25):
    threads.append(threading.Thread(target = job, args = (i, )))
    threads[i].start()
    
for i in range(10):
    print("主程式", i)
    
#print(threads)    
    
for i in threads:
    i.join()
    
print("結束")
```
